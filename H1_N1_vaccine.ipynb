```python
q1 = X_ros['h1n1_worry'].quantile(0.25)
q3 = X_ros['h1n1_worry'].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
import time
```

```python
for i in X_ros.columns :
    q1 = X_ros[i].quantile(0.25)
    q3 = X_ros[i].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - (1.5 * iqr)
    upper = q3 + (1.5 * iqr)
    X_ros[i] = X_ros[i].map(lambda x : lower if x < lower else  upper if x > upper else x )
    print("Done with : ", i)
    time.sleep(1)
```
